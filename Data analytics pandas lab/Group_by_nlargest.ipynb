```python
import pandas as pd
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
df.head()
```

```python
df = df[(df['Year'] > 2013) & (df['Year'] < 2017)]
```

```python
new_df = df.groupby(['Year','Genre']).count().sort_values(['Year'],
        ascending = False).reset_index('Year')
```

```python
new_df.head()
```

```python
new_df.groupby('Year')['Rank'].nlargest(5)
```
